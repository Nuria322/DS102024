## Preparacin y carga de datos

### Libreras

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_iris
```

## Preparacin y carga de datos

```python
data = load_iris()
data.keys()
```

```python
data['data']
```

```python
data['target']
```

```python
data['target_names']
```

```python
data['feature_names']
```

```python
# Creamos el dataframe
variables = pd.DataFrame(data['data'], columns= data['feature_names'])
variables.head (3)
```

```python
# Concateno la matriz con el array 
variables['clase'] = data['target'] # Creo la variables clase y la lleno de los valores de target
variables.head()
```

```python
data['target_names']
```

```python
# Lo podramos hacer as pero solamente cuando son pocos palores que remplazar (en este caso son solo 3)
variables['clase'].replace({0:'setosa', 1:'versicolor', 2: 'virginica'}, inplace = True)
variables.head()
```

```python
# Si tuviesemos ms valores tendriamos que hacerlo asi:
#dict(zip(range(len(data['target_names'])), data['target_names']))
```

```python
# Guardamos el dataframe en una carpeta
variables.to_parquet('Iris_parquet', index = False)
```

```python
df = pd.read_parquet("Iris_parquet")
```

## Limpieza y preprocesado

#### Preprocesado

**Feature Engineer** 
(Crear un dato a partir de datos existentes)

```python
df['sepal area (cm2)'] = df['sepal length (cm)']* df['sepal width (cm)']
df['petal area (cm2)'] = df['petal length (cm)']* df['petal width (cm)']
```

Se crean las columnas de area del petalo y area del spalo

```python
#df = pd.concat([df, clase], axis= 'columns')
#df
```

```python
df = df[['sepal length (cm)', 'sepal width (cm)', 'sepal area (cm2)','petal length (cm)', 'petal width (cm)', 'petal area (cm2)']]
df
```

## Observaciones

1. Se crearon dos variables nuevas que representan el area de cada parte de la flor
