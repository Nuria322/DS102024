### Step 5. Haz un scatter plot con `lmplot()`, que es el método que pinta una regresión lineal
Pasa el DataFrame como argumento del método en el parámetro llamado `data`

```python

%matplotlib inline

# Crear el scatter plot con regresión lineal
sns.lmplot(x='Attack', y='Defense', data=data)

# Añadir título
plt.title('Relación entre Ataque y Defensa de Pokémon')

# Mostrar el gráfico
plt.show()
```

### Step 6. No queremos ver la línea recta de la regresión lineal.
Prueba a añadir fit_reg a falso

```python
# Crear el scatter plot sin la línea de regresión
sns.lmplot(x='Attack', y='Defense', data=data, fit_reg=False)

# Añadir título
plt.title('Relación entre Ataque y Defensa de Pokémon')

# Mostrar el gráfico
plt.show()
```

```python
data_melted = data.melt(var_name='Variable', value_name='Valor', value_vars=data.select_dtypes(include='number').columns)

# Crear el diagrama de cajas
plt.figure(figsize=(12, 6))
sns.boxplot(x='Variable', y='Valor', data=data_melted)

# Añadir título
plt.title('Diagrama de Cajas para Todas las Variables Numéricas')

# Rotar etiquetas del eje x
plt.xticks(rotation=45)

plt.show()
```

### Step 10. No nos interesan las columnas Total, Stage ni Legendary
Bórralos del DataFrame y vuelve a dibujar el diagrama de cajas

```python

data_filtered = data.drop(columns=['Total', '#', 'Legendary'])

data_melted = data_filtered.melt(var_name='Variable', value_name='Valor', value_vars=data_filtered.select_dtypes(include='number').columns)

# Crear el diagrama de cajas
plt.figure(figsize=(12, 6))
sns.boxplot(x='Variable', y='Valor', data=data_melted)
plt.title('Diagrama de Cajas para Todas las Variables Numéricas (sin Total, Stage, ni Legendary)')

# Rotar etiquetas del eje x
plt.xticks(rotation=45)
plt.show()
```
