```python
class Barco:
    def __init__(self, tamano):
        self.tamano = tamano
        self.vidas = tamano

    def recibir_ataque(self):
        self.vidas -= 1
        return self.vidas == 0  # True si el barco está hundido

class Tablero:
    def __init__(self):
        self.tablero = [['O' for _ in range(10)] for _ in range(10)]
        self.barcos = []

    def colocar_barco(self, barco, fila, columna, orientacion):
        # Verificar si la posición y orientación son válidas
        if orientacion not in ['H', 'V']:
            raise ValueError("Orientación inválida. Debe ser 'H' (horizontal) o 'V' (vertical)")
        if not self.es_posicion_valida(barco, fila, columna, orientacion):
            raise ValueError("La posición del barco no es válida")

        # Colocar el barco en el tablero
        for i in range(barco.tamano):
            if orientacion == 'H':
                self.tablero[fila][columna + i] = 'B'
            else:
                self.tablero[fila + i][columna] = 'B'
        self.barcos.append(barco)

    def es_posicion_valida(self, barco, fila, columna, orientacion):
        # Verifica si hay suficiente espacio y si no se superpone con otros barcos
        for i in range(barco.tamano):
            if orientacion == 'H':
                if columna + i >= 10 or self.tablero[fila][columna + i] != 'O':
                    return False
            else:
                if fila + i >= 10 or self.tablero[fila + i][columna] != 'O':
                    return False
        return True

    def recibir_ataque(self, fila, columna):
        if self.tablero[fila][columna] == 'B':
            for barco in self.barcos:
                if self.barco_en_posicion(barco, fila, columna):
                    if barco.recibir_ataque():
                        self.tablero[fila][columna] = 'X'
                        return "Hundido"
                    else:
                        self.tablero[fila][columna] = 'X'
                        return "Tocado"
        else:
            self.tablero[fila][columna] = 'X'
            return "Agua"

    def barco_en_posicion(self, barco, fila, columna):
        # Verifica si un barco ocupa una posición específica
        for i in range(barco.tamano):
            if orientacion == 'H':
                if fila == barco.fila and columna >= barco.columna and columna < barco.columna + barco.tamano:
                    return True
            else:
                if columna == barco.columna and fila >= barco.fila and fila < barco.fila + barco.tamano:
                    return True
        return False

    def mostrar_tablero(self):
        for fila in self.tablero:
            print(' '.join(fila))

    def todos_barcos_hundidos(self):
        for barco in self.barcos:
            if barco.vidas > 0:
                return False
        return True
```
